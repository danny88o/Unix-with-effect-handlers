```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Standar Visulaization scatter plot with line of best fit for dataframe
def plot(df, x, y, title, xlabel, ylabel):
    sns.lmplot(x=x, y=y, data=df)
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.show()
```

```python
# Load from csv and prepare data
def load_from_csv(name, file_path):
    df = pd.read_csv(file_path, index_col='parameter_n')
    df = df[["mean"]]
    df = df.rename(columns={"mean":name})
    df.rename_axis('n', inplace=True)
    return df
```

```python
def concat_dfs(dfs):
    return pd.concat(dfs, axis=1)
```

```python
sync_unison = load_from_csv("Unison", "csv/fib.uc.csv")
sync_ref = load_from_csv("Koka Ref", "csv/sync_ref.kk.out.csv")
sync_param = load_from_csv("Koka Param", "csv/sync_param.kk.out.csv")
sync_state = load_from_csv("Koka State", "csv/sync_state.kk.out.csv")

dfs = [sync_unison, sync_ref, sync_param, sync_state]
df = concat_dfs(dfs)
df.head()
```

```python
df.plot()
plt.savefig("sync.png")
plt.show()
```
